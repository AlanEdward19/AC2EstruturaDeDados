## Exercício 1 

Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
from utils.filaDinamica import Fila
from utils.pilhaDinamica import Pilha


def copiarPilha(pilha : Pilha):
    auxPilha = Pilha()
    returnPilha = Pilha()
    
    while not pilha.isEmpty():
        data = pilha.pop()
        auxPilha.push( data )
    
    while not auxPilha.isEmpty():
        top = auxPilha.pop()

        pilha.push(top)
        returnPilha.push(top)

    return returnPilha

def copiarFila(fila : Fila):
    auxFila = Fila()
    returnFila = Fila()

    while not fila.isEmpty():
        data = fila.dequeue()
        auxFila.enqueue( data )
    
    while not auxFila.isEmpty():
        top = auxFila.dequeue()

        fila.enqueue(top)
        returnFila.enqueue(top)

    return returnFila

def pilhaToFila(pilha: Pilha):
    auxPilha = copiarPilha(pilha)
    auxFila = Fila()

    while not auxPilha.isEmpty():
        auxFila.enqueue(auxPilha.pop())
    
    auxFila = inverterFila(auxFila)

    return auxFila

def inverterFila(fila : Fila):
    auxFila = copiarFila(fila)
    filaInvertida = Fila()
    auxPilha = Pilha()

    while not auxFila.isEmpty():
        auxPilha.push(auxFila.dequeue())

    while not auxPilha.isEmpty():
        filaInvertida.enqueue(auxPilha.pop())
    
    return filaInvertida

def compara(pilha : Pilha, fila : Fila):
    auxPilha = copiarPilha(pilha)
    auxFila = copiarFila(fila)

    encontrouNaOrdem = True
    encontrouInvertido = True

    if auxPilha.size() != auxFila.size:
        return False
    
    pilhaAsFila = pilhaToFila(auxPilha)
    pilhaAsFilaInvertido = inverterFila(pilhaAsFila)

    while not pilhaAsFila.isEmpty():
        if pilhaAsFila.dequeue() != auxFila.dequeue():
            encontrouNaOrdem = False
            break

    auxFila = copiarFila(fila)

    while not pilhaAsFilaInvertido.isEmpty():
        if pilhaAsFilaInvertido.dequeue() != auxFila.dequeue():
            encontrouInvertido = False
            break
    
    return encontrouNaOrdem or encontrouInvertido

##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

# testando a função criada
pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)

comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)

print('Resultado retornado:')
print('\tComparacao 1?: ', comp1)
print('\tComparacao 2?: ', comp2)
print('\tComparacao 3?: ', comp3)
print('\tComparacao 4?: ', comp4)
print('\tComparacao 5?: ', comp5)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tComparacao 1?: True')
print('\tComparacao 2?: True')
print('\tComparacao 3?: True')
print('\tComparacao 4?: True')
print('\tComparacao 5?: False')
````

Faça uma classe chamada **ListaNSE**. Essa classe deve implementar uma **lista não sequencial e estática**. Essa lista deve ter o mesmo comportamento de uma **lista não sequencial e dinâmica**, mas deve ser guardada dentro de uma lista em Python com capacidade limitada.

Essa classe deve possui os seguintes métodos:

- inserir(x, i): insere x na posição i da estrutura
- remover(i): remove o item que está na posição i da estrutura e retorna seu valor 
- buscar(x): buscar o item x na lista e retorna sua posição
- size(): retorna o número de elementos da lista
- clear(): remove os elementos da lista
- \_\_str\_\_(): imprime a estrutura

- Você deve usar a lista do Python apenas para inicializar a estrutura. Depois disso, nenhum método do Python pode ser usado, como append, insert, etc.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class ListaNSE:
    """
    Lista Nao Sequencial Estatica
    """

    def __init__(self, tamanhoMaximo):
        self.tamanhoMaximo = tamanhoMaximo
        self.data = [None]*tamanhoMaximo
        self.numeroDeElementos = 0

    def insert(self, elemento, posicao):

        for i in range(self.numeroDeElementos, posicao, -1):
            self.data[i] = self.data[i-1]

        self.data[posicao] = elemento
        self.numeroDeElementos += 1

    def remove(self, posicao):
        for i in range(posicao, self.numeroDeElementos-1):
            self.data[i] = self.data[i+1]

        self.data[self.numeroDeElementos-1] = None
        self.numeroDeElementos -= 1

    def size(self):
        return self.numeroDeElementos

    def clear(self):
        for i in range(self.numeroDeElementos):
            self.data[i] = None

        self.numeroDeElementos = 0

    def __str__(self):
        count = 0
        strlista = '['
        for i in range(self.numeroDeElementos):
            strlista += str(self.data[i])
            count += 1
            if count == self.numeroDeElementos:
                break
            strlista += ','
        strlista += ']'
        return strlista



##########################################################################

# testa a Lista
lista = ListaNSE(5)        

# inserir no inicio da lista
lista.insert(5,0)
print( 'insert(5,0): ', lista )

lista.insert(10,0)
print( 'insert(10,0): ', lista )

lista.insert(24,2)
print( 'insert(24,2): ', lista )

lista.insert(25,1)
print( 'insert(25,1): ', lista )

lista.remove(2)
print( 'remove(2): ', lista )

print( 'Size: ', lista.size() )

lista.clear()
print( 'Clear: ', lista )

print('\n' + 20*'-')
print('Resultado esperado:')
print( 'insertHead(5,0): [5]')
print( 'insertHead(10,0): [10,5]')
print( 'insert(24,2): [10,5,24]')
print( 'insert(25,1): [10,25,5,24]')
print( 'remove(2): [10,25,24]')
print( 'Size: 3')
print( 'Clear: []')
````

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    """
    Noh de uma lista duplamente ligada
    """
    def __init__(self, data = None, next = None, prev = None):
        self.data = data
        self.next = next # apontador para o noh da frente
        self.prev = prev # apontador para o noh de tras

class ListaInsertion:
    """
    Lista duplamente ligada
    """

    def __init__(self, data  = None):
        self.head = None
        self.tail = None
        self.size = 0

    def insert(self, data):
        """
        Adiciona um elemento no inicio da lista
        """

        # cria um novo noh com o dado informado
        tempNode = Node(data)

        # se a lista esta vazia, o element head
        # ira receber o novo noh.
        # Nesse caso,o elemento tail devera apontar para
        # o mesmo noh do elemento head
        if self.size==0:
            self.head = tempNode
            self.tail = tempNode

        else:

            # o proximo elemento na lista a frente do novo noh sera o noh head antigo
            tempNode.next = self.head

            # o elemento atras do noh head antigo sera o novo noh
            self.head.prev = tempNode

            # o novo noh head sera o novo noh
            self.head = tempNode

        # incrementa a quantidade de elementos da lista
        self.size += 1

    def insertTail(self, data):
        """
        Adiciona um elemento no final da lista
        """

        # cria um novo noh com o dado informado
        tempNode = Node(data)

        # se a lista esta vazia, o element tail
        # ira receber o novo noh.
        # Nesse caso,o elemento head devera apontar para
        # o mesmo noh do elemento tail
        if self.size==0:
            self.tail = tempNode
            self.head = self.tail

        else:

            # o tail antigo ficara atras do novo noh e, por isso,
            # o apontador prev do novo noh deve apontar para o tail antigo
            tempNode.prev = self.tail

            # o noh tail sera substituido pelo novo noh e, por isso,
            # seu apontador next deve apontar para o novo noh
            self.tail.next = tempNode

            # o novo noh tail sera o novo noh
            self.tail = tempNode

        self.size += 1

    def busca(self, data):
        """
        Busca um noh com o valor informado

        Retorna a posicao do valor ou -1 caso nao exista
        """

        if self.size > 0:

            # comecao pelo head
            tempNode  = self.head
            auxValor = tempNode.data

            if auxValor == data:
                return 0

            # percorre o resto da lista ate que o apontador next do noh seja nulo
            # o que indica que chegou no tail
            idx = 1
            while tempNode.next is not None:

                # vai para o proximo noh
                tempNode = tempNode.next

                auxValor = tempNode.data

                if auxValor == data:
                    return idx

                idx += 1

        # se chegar neste return significa que o valor nao foi encontrado
        return -1


    def insertPosition(self, data, position):
        """
        Adiciona um elemento na posição especificada
        """

        if position >= self.size or position < 0:
            print( 'Erro! A posicao desejada nao existe na lista' )

        else:

            # cria um novo noh com o dado informado
            tempNode = Node(data)

            # se a lista esta vazia, o element tail
            # ira receber o novo noh.
            # Nesse caso,o elemento head devera apontar para
            # o mesmo noh do elemento tail
            if self.size==0:
                self.tail = tempNode
                self.head = self.tail

            else:

                # percorre a lista ate encontrar o noh da posicao anterior a desejada.
                # Comeca pelo head
                prevNode = self.head

                idx = 1
                while (idx < position):

                    # vai para o proximo noh
                    prevNode = prevNode.next

                    idx += 1

                # o apontador next do novo noh deve apontar para o noh que esta na frente do prevNode
                tempNode.next = prevNode.next

                # o apontador prev do novo noh deve apontar para o prevNode
                tempNode.prev = prevNode

                # o apontador prev do noh indicado pelo apontador next do prevNode
                # devera apontar para o novo noh
                prevNode.next.prev = tempNode

                # o apontador next do prevNode deve apontar para o novo noh
                prevNode.next = tempNode

            self.size += 1

    def getSize(self):
        """
        Retorna a quantidade de elementos da lista
        """

        return self.size

    def isEmpty(self):
        """
        Retorna True se a lista esta vazia
        """

        return self.head is None

    def remove(self, position):
        """
        Remove um elemento da lista
        """

        if position >= self.size:
            print( 'Erro! A posicao desejada nao existe na lista' )

        else:

            # se a lista esta vazia, retorna um erro
            if self.size==0:
                print('Erro. A lista esta vazia (underflow')
                return False

            else:

                # percorre a lista ate encontrar o noh da posicao desejada.
                # Comeca pelo head
                tempNode = self.head

                idx = 1
                while (idx <= position):

                    # vai para o proximo noh
                    tempNode = tempNode.next

                    idx += 1

                # se o noh estiver na primeira posicao da lista
                # o head recebe o next do noh removido
                if position == 0:
                    self.head = tempNode.next

                # se o noh nao estiver na primeira posicao da lista,
                # o apontador next do noh anterior ao que sera removido deve apontar para o
                # o noh indicado pelo apontador next do noh que sera removido
                else:
                    tempNode.prev.next = tempNode.next


                # se o noh estiver na ultima posicao da lista
                # o tail recebe o prev do noh removido
                if position == self.size - 1:
                    self.tail = tempNode.prev

                # se o noh nao estiver na ultima posicao da lista,
                # o apontador prev do noh da frente do que sera removido deve apontar para o
                # o noh indicado pelo apontador prev do noh que sera removido
                else:
                    tempNode.next.prev = tempNode.prev

                # deleta o noh da memoria
                del tempNode

            self.size -= 1


    def clear(self):
        """
        Limpa a lista
        """

        while not self.isEmpty():
            self.remove(0)

    def __str__(self):
        """
        Retorna uma string com as informacoes da lista
        quando o objeto e chamado dentro de um print()
        ou dentro de um str().

        A lista nao possui restricao de acesso aos seus elementos.
        Por isso, ela nao precisa ser destruida para que seus elementos sejam
        visitados como na fila e na pilha
        """

        strfila = '[ '

        if self.size > 0:

            # comecao pelo head
            tempNode  = self.head
            valor = tempNode.data

            # guarda o valor do head em uma string
            strfila += ' ' + str(valor)

            # percorre o resto da lista ate que o apontador next do noh seja nulo
            # o que indica que chegou no tail
            while tempNode.next is not None:

                # vai para o proximo noh
                tempNode = tempNode.next

                valor = tempNode.data

                # guarda o valor em uma string
                strfila += ' ' + str(valor)

        strfila += ' ]'

        return strfila

    def binarySearch(self, value, left, right, list):
        middle = (left+right)//2
        if list[middle]>value:
            return self.binarySearch(value, left, middle-1, list)

        elif list[middle]<value:
            return self.binarySearch(value, middle+1, right, list)

        else:
            return  middle



    def getCharMean(self, string):
        stringList = string.split()
        charTotal = 0

        for i in range(len(stringList)-1):
            charTotal += len(stringList[i])

        return charTotal // len(stringList)

    def comparar(self, string1, string2):
        return self.getCharMean(string1) == self.getCharMean(string2)


##########################################################################

# testa o exercicio
lista = ListaInsertion()

# inserir no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n", lista)

qtd1 = lista.repeticoes(3)
qtd2 = lista.repeticoes(2)
qtd3 = lista.repeticoes(30)
print("Qtd. repeticoes 1: ", qtd1)
print("Qtd. repeticoes 2: ", qtd2)
print("Qtd. repeticoes 3: ", qtd3)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, A menina gosta de jogar, ', end="")
print('O menino gosta de andar, ', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")
print('Centro Universitario Facens]', end="")
````

Modifique a árvore binária de busca ensinada nesta disciplina e dê o nome de **TreePath**. Acrescente nessa árvore um método chamado **caminhos**. Esse método deve preencher um objeto da lista duplamente ligada fornecida junto com este notebook com a soma do valores dos nós que estão no caminho de cada um dos nós folhas até a raiz da árvore. Por exemplo, considere a árvore apresentada abaixo:

```
            ___________61___________
           |                        |
       ____34_____              ____67____
      |           |            |          |    
      19__        50__      __66        __94__
          |           |    |           |      |  
         20           55   64          70     95
```

Se a árvore acima fosse apresentada como entrada, o método **caminhos** deveria retornar uma lista duplamente ligada com os seguintes valores: [134, 200, 258, 292, 317], que é formada pelas seguites somas: [61+34+19+20, 61+34+50+55, 61+67+66+64, 61+67+94+70, 61+67+94+95].

````python

################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None

class NodeDuplamenteLigado:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None


class ListaDuplamenteLigada:
    def __init__(self):
        self.head = NodeDuplamenteLigado()
        self.tail = NodeDuplamenteLigado()
        self.head.next = self.tail
        self.tail.prev = self.head

    def adicionar(self, data):
        new_node = NodeDuplamenteLigado(data)
        new_node.prev = self.tail.prev
        new_node.next = self.tail
        self.tail.prev.next = new_node
        self.tail.prev = new_node

    def _join(self, delimitador, lista):
        
        if len(lista) == 0:
            
            return ""
    
        resultado = str(lista[0])
    
        for i in lista[1:]:

            resultado += delimitador + str(i)
    
        return resultado

class TreePath:
    def __init__(self):
        self.root = None

    def insertRecursivo(self, data, node):
        if data < node.data:
            
            if node.left is None:
                node.left = Node(data)
            else:
                self.insertRecursivo(data, node.left)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self.insertRecursivo(data, node.right)

    def caminhosRecursivo(self, node, lista, total):
        if node is None:
            return
        total += node.data
        if node.left is None and node.right is None:
            lista.adicionar(total)
            return
        self.caminhosRecursivo(node.left, lista, total)
        self.caminhosRecursivo(node.right, lista, total)

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self.insertRecursivo(data, self.root)

    def caminhos(self):
        lista = ListaDuplamenteLigada()
        self.caminhosRecursivo(self.root, lista, 0)

        node = lista.head.next

        result = []

        while node != lista.tail:

            result[len(result):] = [str(node.data)]

            node = node.next

        resultList = lista._join(", ",result)

        return f"[{resultList}]"
        


##########################################################################

tree1 = TreePath()
tree1.insert(61)
tree1.insert(34)
tree1.insert(67)
tree1.insert(19)
tree1.insert(50)
tree1.insert(66)
tree1.insert(94)
tree1.insert(20)
tree1.insert(55)
tree1.insert(64)
tree1.insert(70)
tree1.insert(95)

lista = tree1.caminhos()
print(lista)

print('\n' + 20*'-')
print('Resultado esperado:')
print("[134, 200, 258, 292, 317]")
````
